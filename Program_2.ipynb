```python
seq = input("Enter DNA sequence: ").upper()

start_codon = "ATG"
stop_codons = ["TAA", "TAG", "TGA"]

start_index = -1
for i in range(len(seq) - 2):
    if seq[i:i+3] == start_codon:
        start_index = i
        break

if start_index != -1:
    stop_index = -1
    found_stop = ""

    for i in range(start_index, len(seq) - 2, 3):
        codon = seq[i:i+3]
        if codon == "TAA" or codon == "TAG" or codon == "TGA":
            stop_index = i
            found_stop = codon
            break

    print(f"Start Codon (ATG) at index: {start_index}")
    if stop_index != -1:
        print(f"Stop Codon ({found_stop}) at index: {stop_index}")
        print(f"CDS: {seq[start_index:stop_index+3]}")
    else:
        print("No in-frame stop found.")
else:
    print("No start codon found.")
```

```python
seq1 = input("Enter first DNA sequence: ").upper()
seq2 = input("Enter second DNA sequence: ").upper()

distance = 0

if len(seq1) == len(seq2):
    for i in range(len(seq1)):
        if seq1[i] != seq2[i]:
            distance += 1

    print(f"Sequence 1: {seq1}")
    print(f"Sequence 2: {seq2}")
    print(f"Hamming Distance: {distance}")
else:
    print("Error: Sequences must be of equal length for Hamming Distance.")
```
